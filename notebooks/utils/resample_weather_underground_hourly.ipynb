```python
import pandas as pd
```

```python
input_dir_name = '../../data/weather/weather_underground/KDCA/'
```

```python
file_name = 'weather_underground_processed_total.csv'
```

```python
df = pd.read_csv(input_dir_name + '{}'.format(file_name))
```

```python
df
```

```python
datetime_series = pd.to_datetime(df['DateTime'])
datetime_index = pd.DatetimeIndex(datetime_series.values)
df2 = df.set_index(datetime_index)
df2.drop('DateTime',axis=1,inplace=True)
```

```python
result = df2.Temperature.resample('30min').mean().interpolate(method='linear')
```

```python
result.to_csv(input_dir_name + '{}.csv'.format('processed'), index=True)
```
